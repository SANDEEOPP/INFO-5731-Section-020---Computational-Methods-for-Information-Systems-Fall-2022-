Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup
import pandas as pd
main_text = [] # List to store Review headings
sub_text =[] #List to store reviews

#Collected the Nike Revolution 5 Running shoew review page link
for number in range(100):
  requested_link = "https://www.flipkart.com/nike-revolution-5-running-shoes-men/product-reviews/itm11637de31a718?pid=SHOFK455USFFYRMG&lid=LSTSHOFK455USFFYRMGWHZANI&marketplace=FLIPKART" + str(number) # Generating link dynamically
  #Getting Request for the link above
  web_page = requests.get(requested_link)
  soup = BeautifulSoup(web_page.text, 'html.parser')
  rating_reviews = soup.find_all(class_='_3LWZlK _1BLPMq _3B8WaH')
  text_reviews = soup.find_all(class_='col _2wzgFH K0kLPL _1QgsS5')
  for ele, sub_ele in zip(rating_reviews, text_reviews) : # Iterating through the list
      main_text.append(ele.text) #Appending to empty list
      sub_text.append(sub_ele.text)

#Dataframe Creation by combining collected data and making the column names of our own!
Nike_Reviews = pd.DataFrame(list(zip(main_text, sub_text)), columns =['Rating', 'Glimpse of Review']) 
print("Obtained Data Frame Length is {0}".format(len(Nike_Reviews)))
Nike_Reviews
```

```python
# You code here (Please add comments in the code):
from bs4 import BeautifulSoup#importing python packag Beautifulsoup
import requests, lxml, os, json
import pandas as pd#importing pandas
headers = {
    'User-agent':
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582"
}
params = {
  "q": "Information Science",
  "hl": "en",
}
google_link = requests.get('https://scholar.google.com/scholar', headers=headers, params = params).text#here we did for google scholar
soup = BeautifulSoup(google_link, 'lxml')
for pdf in soup.select('.gs_or_ggsm a'):
  pdf_file_link = pdf['href']
  print(pdf_file_link)
  data = []
for result in soup.select('.gs_ri'):
  T_header = result.select_one('.gs_rt').text
  publication_info = result.select_one('.gs_a').text
  abstract = result.select_one('.gs_rs').text
  try:
    version_of_articles = result.select_one('a~ a+ .gs_nph')['href']
  except:
    version_of_articles = None
  d = {
      'TITLE': T_header,'Authors,Venue,journal,conference,Year': publication_info,'ABSTRACT': abstract
  }
  data.append(d)
data_frame = pd.DataFrame(data)
data_frame.head()
#reff https://dev.to/dmitryzub/scrape-google-scholar-with-python-32oh
```
