```python
# You code here (Please add comments in the code):
import pandas as pd

text_file="http://www.textfiles.com/adventure/abyssal.txt"
text=pd.read_csv(text_file,error_bad_lines=False,names=['lines'])
text.head(10)

```

```python
# Num of sentences

text['Num of Sentences'] = text['lines'].apply(lambda x: len(str(x).split(".")))

text[['lines','Num of Sentences']]
```

```python
# Num of Words

text['No of Words']=text['lines'].apply(lambda x: len(x.split(" ")))

text[['lines','No of Words']]
```

```python
# Num of Characters

text['Num of Characters'] = text['lines'].str.len()
text[['lines','Num of Characters']]
```

```python
# Num of Stopwords
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

text['Num of Stopwords'] = text['lines'].apply(lambda x: len([x for x in x.split() if x in stop]))
text[['lines','Num of Stopwords']]
```

```python
# Num of Special Characters

text['special character count']=text['lines'].apply(lambda x: len([y for y in x.split() if y.startswith(('#','@'))]))
text[['lines','special character count']]
```

```python
# Num of Numerics

text['numeric count']=text['lines'].apply(lambda x: len([y for y in x.split() if y.isdigit()]))
text[['lines','numeric count']]
```

```python
# Num of uppercase words

text['uppercase words']=text['lines'].apply(lambda x: len([y for y in x.split() if y.isupper()]))
text[['lines','uppercase words']]
```

```python
# Lower casing

text['lines']=text['lines'].str.lower()
text['lines']
```

```python
# Punctuation removal

import string
text['lines']=text['lines'].str.replace('[{}]'.format(string.punctuation), '')
text['lines']
```

```python
# Stopwords removal

from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')

text['lines']=text['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))
text['lines']
```

```python
# Frequent words removal

freq=pd.Series(' '.join(text['lines']).split()).value_counts()[:10]
freq=list(freq.index)

text['lines']=text['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in freq))
text['lines']
```

```python
# Rare words removal

rare=pd.Series(' '.join(text['lines']).split()).value_counts()[-10:]

rare=list(rare.index)
text['lines']=text['lines'].apply(lambda x:" ".join(x for x in x.split() if x not in rare))
text['lines']
```

```python
# Spelling correction

from textblob import TextBlob

text['lines']=text['lines'].apply(lambda x: str(TextBlob(x).correct()))
text['lines']
```

```python
# Stemming

from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

text['lines']=text['lines'].apply(lambda x: " ".join([stemmer.stem(word) for word in x.split()]))
text['lines']
```

```python
# You code here (Please add comments in the code):


term_freq = (text['lines'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
term_freq.columns = ['words','term_freq']
term_freq
```

```python
import numpy as np

for i,word in enumerate(term_freq['words']):
    term_freq.loc[i, 'inverse doc freq'] = np.log(text.shape[0]/(len(text[text['lines'].str.contains(word)])))

term_freq
```

```python
#TF-IDF

term_freq['TF-IDF'] = term_freq['term_freq'] * term_freq['inverse doc freq']
term_freq
```
